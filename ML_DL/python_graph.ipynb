```python
# 여러개의 선 그리기
t = np.arange(0., 5., 0.2)
print(t)
plt.figure(figsize=(10, 6))
plt.title('line plot에서 여러개의 선 그리기')
plt.plot(t, t, 'r--') # red, dashed line
plt.plot(t, 0.5*t**2, 'bs:') # blue, square marker, dot line style
plt.plot(t, 0.2*t**3, 'g^-') # green, triangle_up marker, solid line style
plt.show()
```

```python
# 위 코드를 한 줄로 표현
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
```

```python
# 여러개의 선 그리기(plot 한 번 사용)
t = np.arange(0., 5., 0.2)
print(t)
plt.figure(figsize=(10, 6))
plt.title('line plot에서 여러개의 선 그리기: plot 한 번 사용')
plt.plot(t, t, 'r--', t, 0.5*t**2, 'bs:', t, 0.2*t**3, 'g^-')
plt.show()
```

## Bar Chart

```python
y = [2, 3, 1, 4]
x = np.arange(len(y))
z = [2, 3]
s = [0, 1]

e = [1, 4]
h = [2, 3]

xlabel = ['가', '나', '다', '라']
plt.figure(figsize=(10, 6))
plt.title('Bar Chart')
plt.bar(x, y)
plt.plot(s, z, 'g--o')
plt.plot(h, e, 'b--o')
plt.xticks(x)
plt.yticks(sorted(y))
plt.text(-0.1, 1, r'test')
plt.xlabel('가나다라')
plt.ylabel('빈도수')
print(x)
plt.show()
```

```python
np.random.seed(0)
people=['몽룡', '춘향', '방자', '향단']
y = np.arange(len(people))
performance = 3 + 10 * np.random.rand(len(people))
print(y)
print(performance)

plt.title('Bar Chart')
plt.barh(y, performance, alpha=0.4) # 주의: 첫 번째 데이터 y가 y축으로 표현됨
plt.yticks(y, people)
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('dark')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
plt.show()
```

```python
# 축선 표시 여부를 결정하는 함수
x = np.linspace(1, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x)
y3 = 3*np.sin(x)
y4 = 4*np.sin(x)

sns.set_style('whitegrid')
plt.figure(figsize=(10, 6))
plt.plot(x, y1, x, y2, x, y3, x, y4)
sns.despine(right=False, left=False) # top=True, bottom=False default
plt.show()
```

## 단계 구분도

```python
# 데이터를 지도 그림에 반영시켜 전달하는 그래프
# folium에 layer로 얹게 된다.

state_unemployment = 'data/02. folium_US_Unemployment_Oct2012.csv'
state_unemp_data = pd.read_csv(state_unemployment)
state_unemp_data.head()
```

```python
state_geo = 'data/02. folium_us-states.json'
```

```python
map = folium.Map(location=[40, -98], zoom_start=3)
map.choropleth(geo_data=state_geo,
               data=state_unemp_data,
               columns=['State', 'Unemployment'],
               key_on='feature.id',
               fill_color='YlGn',
               legend_name='Unemployment Rate(%)')
folium.LayerControl().add_to(map)
map
```

```python
# 한국 지도 데이터 이용해 단계구분도 그리기
import json

geo_path='data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

map=folium.Map(location=[37.5502, 126.982],
               zoom_start=11,
               tiles='Stamen Toner')

f_test = pd.read_csv('data/cctv_result.csv', encoding='utf-8', index_col='구별')
f_test.head()
```

```python
map.choropleth(geo_data=geo_str,
               data=f_test,
               columns=[f_test.index, '소계'],
               key_on='feature.id',
               fill_color='PuRd',
               legend_name='CCTV 현황')
folium.LayerControl().add_to(map)
map
```

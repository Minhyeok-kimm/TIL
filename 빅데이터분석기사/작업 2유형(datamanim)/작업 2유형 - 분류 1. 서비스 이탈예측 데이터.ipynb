## 데이터 불러오기

```python
import pandas as pd
#데이터 로드
x_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_train.csv")
y_train = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/y_train.csv")
x_test= pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/churnk/X_test.csv")


display(x_train.head())
display(y_train.head())
display(x_test.head())
```

## 고객의 신상정보 데이터를 통한 회사 서비스 이탈 예측(종속변수: Exited)

### 데이터 정보 확인

```python
print(x_train.info())
```

```python
print(x_train.isna().sum().sum(), y_train.isna().sum().sum(), x_test.isna().sum().sum(), sep='\n') # Null값 없음
```

```python
print(x_train.shape, x_test.shape)
```

### encoding

```python
x_train.drop(columns=['CustomerId', 'Surname'], inplace=True)
id = x_test.CustomerId
x_test.drop(columns=['CustomerId', 'Surname'], inplace=True)
x_train['Gender'] = x_train['Gender'].astype('category').cat.codes
x_test['Gender'] = x_test['Gender'].astype('category').cat.codes
x_train['Geography'] = x_train['Geography'].astype('category').cat.codes
x_test['Geography'] = x_test['Geography'].astype('category').cat.codes

print(x_test.info())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

ls = ['CreditScore','Age', 'Balance','EstimatedSalary','Tenure','NumOfProducts','HasCrCard','IsActiveMember','EstimatedSalary']
x_train[ls] = scaler.fit_transform(x_train[ls])
x_test[ls] = scaler.fit_transform(x_test[ls])
```

```python
display(x_train.head())
display(y_train.head())
display(x_test.head())
```

```python
y_train.drop(columns=['CustomerId'], inplace=True)
```

### 검증용 데이터 분리

```python
from sklearn.model_selection import train_test_split
x_tr, x_ts, y_tr, y_ts = train_test_split(x_train, y_train, test_size=0.3)
```

### 모델링

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(random_state=1234)

model.fit(x_tr, y_tr['Exited'])
print(model.score(x_tr, y_tr), model.score(x_ts, y_ts))
```

### 예측

```python
pred = model.predict(x_test)
# print(id)
pred = pd.Series(pred, name='Exited')
# print(pred)
final = pd.concat([id, pred], axis=1)
print(final)
```

```python
final.to_csv('서비스이탈예측.csv', index=False)
```

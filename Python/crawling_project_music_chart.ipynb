```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
import time
```

```python
melon_url = 'https://www.melon.com/chart/index.htm'
bugs_url = 'https://music.bugs.co.kr/chart'
ginie_url = 'https://www.genie.co.kr/chart/top200'
```

```python
# 브라우저 실행 후 멜론 차트 페이지 접속
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get(melon_url)
```

```python
# 곡 title 찾기(50위까지)
m_t50_t = driver.find_elements(By.CSS_SELECTOR, '#conts .lst50 .rank01')
print(len(m_t50_t))
for i in m_t50_t:
    print(i.text)
```

```python
# 가수 이름 찾기(50위까지)
m_t50_s = driver.find_elements(By.CSS_SELECTOR, '#conts .lst50 .rank02')
print(len(m_t50_s))
for i in m_t50_s:
    print(i.text)
```

```python
# 51~100위 곡명, 가수명 추출
m_t100_t = driver.find_elements(By.CSS_SELECTOR, '#conts .lst100 .rank01')
m_t100_s = driver.find_elements(By.CSS_SELECTOR, '#conts .lst100 .rank02')
```

```python
# 멜론차트 곡, 가수명 리스트 생성
m_t = []
m_s = []

# 곡명, 가수명 데이터만 추출하는 함수 생성
def split_data(lst, data):
    for i in data:
        lst.append(i.text)
```

```python
# 함수 실행 후 결과 확인
split_data(m_t, m_t50_t)
split_data(m_t, m_t100_t)
split_data(m_s, m_t50_s)
split_data(m_s, m_t100_s)
print(len(m_t), len(m_s))
```

```python
# 날짜, 시간 추출
day = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .year').text
time = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .hour').text
print(day, time)
```

```python
# 멜론 차트 정보 생성
melon_df = pd.DataFrame({'date':day+' '+time, 'site_name':'melon' ,'title':m_t, 'singer':m_s})
print(melon_df.head())
```

```python
# 차트 정보 저장
melon_df.to_csv('./crawling/melon_chart.csv', index=False)
```

```python
# 멜론 차트 정보 크롤링
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
import time

melon_url = 'https://www.melon.com/chart/index.htm'
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get(melon_url)

m_t100_t = driver.find_elements(By.CSS_SELECTOR, 'div.rank01 a')
m_t100_s = driver.find_elements(By.CSS_SELECTOR, 'div.rank02')

m_t = []
m_s = []

def split_data(lst, data):
    for i in data:
        lst.append(i.text)

split_data(m_t, m_t100_t)
split_data(m_s, m_t100_s)

day = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .year').text
now_time = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .hour').text

melon_df = pd.DataFrame({'date':day+' '+now_time, 'site_name':'melon', 'rank':[i for i in range(1, len(m_t)+1, 1)], 'title':m_t, 'singer':m_s})

from datetime import datetime
path = './crawling/'
file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
file_path = path + 'melon_chart' + file_name

melon_df.to_csv('./crawling/melon_chart.csv', index=False)

driver.close()
```

```python
# 벅스 차트 정보 크롤링
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
import time

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
bugs_url = 'https://music.bugs.co.kr/chart'
driver.get(bugs_url)
b_t100_t = driver.find_elements(By.CSS_SELECTOR, 'table.byChart p.title a')
b_t100_s = driver.find_elements(By.CSS_SELECTOR, 'table.byChart p.artist a:not(.more)')

b_t = []
b_s = []

for i in b_t100_t:
    b_t.append(i.text)

for i in b_t100_s:
    b_s.append(i.text)

date = driver.find_element(By.CSS_SELECTOR, '.filterChart time').text
day_lst = date.split(' ')
day = day_lst[0]
time = day_lst[1]

bugs_df = pd.DataFrame({'date':day+' '+time, 'site_name':'bugs', 'rank':[i for i in range(1, len(b_t)+1, 1)], 'title':b_t, 'singer':b_s})

from datetime import datetime
path = './crawling/'
file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
file_path = path + 'bugs_chart' + file_name

bugs_df.to_csv(file_path, index=False)
    
driver.close()
```

```python
# 지니 차트 정보 크롤링
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import numpy as np
import pandas as pd
import time

ginie_url = 'https://www.genie.co.kr/chart/top200'

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
driver.get(ginie_url)
g_t50_t = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.title')
g_t50_s = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.artist')

g_t = []
g_s = []

for i in g_t50_t:
    g_t.append(i.text)
    
for i in g_t50_s:
    g_s.append(i.text)

time.sleep(3)
pages = driver.find_elements(By.CSS_SELECTOR, 'div.page-nav a')
pages[1].click()

time.sleep(3)
g_t100_t = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.title')
g_t100_s = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.artist')

for i in g_t100_t:
    g_t.append(i.text)

for i in g_t100_s:
    g_s.append(i.text)

day = driver.find_element(By.CSS_SELECTOR, 'div.date span#inc_date').text
now_time = driver.find_element(By.CSS_SELECTOR, 'div.date span#inc_time').text

ginie_df = pd.DataFrame({'date':day + ' ' + now_time, 'site_name':'ginie', 'rank':[i for i in range(1, len(g_t)+1, 1)], 'title':g_t, 'singer':g_s})

from datetime import datetime
path = './crawling/'
file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
file_path = path + 'ginie_chart' + file_name

ginie_df.to_csv(file_path, index=False)

driver.close()
```

```python
# 멜론 앨범명 추가
def melon():
    melon_url = 'https://www.melon.com/chart/index.htm'
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    driver.get(melon_url)

    m_t100_t = driver.find_elements(By.CSS_SELECTOR, 'div.rank01 a')
    m_t100_s = driver.find_elements(By.CSS_SELECTOR, 'div.rank02')
    m_t100_a = driver.find_elements(By.CSS_SELECTOR, 'div.rank03 a')

    m_t = []
    m_s = []
    m_a = []

    def split_data(lst, data):
        for i in data:
            lst.append(i.text)

    split_data(m_t, m_t100_t)
    split_data(m_s, m_t100_s)
    split_data(m_a, m_t100_a)

    day = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .year').text
    now_time = driver.find_element(By.CSS_SELECTOR, '#conts .calendar_prid .hour').text

    melon_df = pd.DataFrame({'date':day+' '+now_time, 'site_name':'melon', 'rank':[i for i in range(1, len(m_t)+1, 1)], 'title':m_t, 'singer':m_s, 'album':m_a})

    from datetime import datetime
    path = './crawling/'
    file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
    file_path = path + 'melon_chart' + file_name

    melon_df.to_csv(file_path, index=False)
    
    print(melon_df)

    driver.close()
```

```python
# 벅스 앨범명 추가
def bugs():
    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    bugs_url = 'https://music.bugs.co.kr/chart'
    driver.get(bugs_url)
    b_t100_t = driver.find_elements(By.CSS_SELECTOR, 'table.byChart p.title a')
    b_t100_s = driver.find_elements(By.CSS_SELECTOR, 'table.byChart p.artist a:not(.more)')
    b_t100_a = driver.find_elements(By.CSS_SELECTOR, 'table.byChart a.album')

    b_t = []
    b_s = []
    b_a = []

    for i in b_t100_t:
        b_t.append(i.text)

    for i in b_t100_s:
        b_s.append(i.text)

    for i in b_t100_a:
        b_a.append(i.text)

    date = driver.find_element(By.CSS_SELECTOR, '.filterChart time').text
    day_lst = date.split(' ')
    day = day_lst[0]
    time = day_lst[1]

    bugs_df = pd.DataFrame({'date':day+' '+time, 'site_name':'bugs', 'rank':[i for i in range(1, len(b_t)+1, 1)], 'title':b_t, 'singer':b_s, 'album':b_a})

    from datetime import datetime
    path = './crawling/'
    file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
    file_path = path + 'bugs_chart' + file_name

    bugs_df.to_csv(file_path, index=False)
    
    print(bugs_df)
        
    driver.close()
```

```python
# 지니 앨범명 추가
def ginie():
    ginie_url = 'https://www.genie.co.kr/chart/top200'

    driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
    driver.get(ginie_url)
    g_t50_t = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.title')
    g_t50_s = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.artist')
    g_t50_a = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.albumtitle')

    g_t = []
    g_s = []
    g_a = []

    for i in g_t50_t:
        g_t.append(i.text)
    
    for i in g_t50_s:
        g_s.append(i.text)
    
    for i in g_t50_a:
        g_a.append(i.text)

    time.sleep(3)
    pages = driver.find_elements(By.CSS_SELECTOR, 'div.page-nav a')
    pages[1].click()

    time.sleep(3)
    g_t100_t = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.title')
    g_t100_s = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.artist')
    g_t100_a = driver.find_elements(By.CSS_SELECTOR, 'table.list-wrap a.albumtitle')

    for i in g_t100_t:
        g_t.append(i.text)

    for i in g_t100_s:
        g_s.append(i.text)

    for i in g_t100_a:
        g_a.append(i.text)

    day = driver.find_element(By.CSS_SELECTOR, 'div.date span#inc_date').text
    now_time = driver.find_element(By.CSS_SELECTOR, 'div.date span#inc_time').text

    ginie_df = pd.DataFrame({'date':day + ' ' + now_time, 'site_name':'ginie', 'rank':[i for i in range(1, len(g_t)+1, 1)], 'title':g_t, 'singer':g_s, 'album':g_a})

    from datetime import datetime
    path = './crawling/'
    file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
    file_path = path + 'ginie_chart' + file_name

    ginie_df.to_csv(file_path, index=False)
    
    print(ginie_df)

    driver.close()
```

```python
# 지정 사이트에 따라 달라지는 top_100 함수 생성하기
def top_100(site):
    from selenium import webdriver
    from selenium.webdriver.chrome.service import Service as ChromeService
    from webdriver_manager.chrome import ChromeDriverManager
    from selenium.webdriver.common.by import By
    import numpy as np
    import pandas as pd
    import time
    if site.lower() == 'melon':
        melon()
    elif site.lower() == 'bugs':
        bugs()
    elif site.lower() == 'ginie':
        ginie()
    else:
        print('올바른 사이트를 입력하세요')

if __name__ == '__main__':
    site = input('사이트를 입력하세요(melon, bugs, ginie 지원)')
    top_100(site)
```

```python
# 통합 자료 함수 만들기
def total_chart():
    import time
    melon()
    time.sleep(3)
    bugs()
    time.sleep(3)
    ginie()
    
    total = pd.concat([melon_df, bugs_df, ginie_df])

    from datetime import datetime
    path = './crawling/'
    file_name = datetime.now().strftime('%Y%m%d%H') + '.csv'
    file_path = path + file_name

    total.to_csv(file_path, index=False)

if __name__ == '__main__':
    total_chart()
```

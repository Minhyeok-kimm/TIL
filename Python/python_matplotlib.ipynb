## 다중 그래프

```python
# 가) fig.add_subplot(행,열, n번째열)
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)  # 1행 2열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

ax2= fig.add_subplot(1,2,2)   # 1행 2열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub2")

fig.suptitle("Main")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,(1, 2))  # 1행 1열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

# ax2= fig.add_subplot(2,2,2)   # 1행 2열
# plt.plot([-1,-2,-3,-4])
# plt.title("sub2")

ax3 = fig.add_subplot(2,2,3)  # 2행 1열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub3")

ax4= fig.add_subplot(2,2,4)   # 2행 2열
plt.plot([-1,-2,-3,-4])
plt.title("sub4")

fig.suptitle("Main")
plt.show()
```

```python
# 나) plt.subplots(행,열)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=2) # 1행 2열
print(axs)
axs[0].plot([1,2,3])
axs[1].plot([-1,-2,-3])
plt.show()
```

```python
# 10) grid
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,2, 0.2)
plt.grid(True, axis='both', color='blue', linestyle=':') # axis = {'both', 'x', 'y'}
plt.plot(n,n, 'bo', label='linear')
plt.plot(n,n**2, 'r', marker="*", linewidth=2, label='quadratic')
plt.plot(n,n**3, 'y', marker="^", linewidth=3, label='cubic')
plt.legend()
plt.show()
```

## 막대그래프
- 범주형의 값을 직사각형의 막대로 표현
- [공식레퍼런스](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar)

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(3)
values = [100, 400, 900]

plt.bar(x, values) # x, height
# plt.axhline(200, color='r', linestyle="--")
# plt.axvline(0.5, color='k')  # x=0.5 값에  수직 line 생성
plt.xticks(x, labels=['2017', '2018', '2019'])
plt.show()
```

```python
x = np.arange(3)
values = [100, 400, 900]

plt.bar(x, values, color='green', edgecolor='red', alpha=0.3, linestyle='--', width=0.4, align='edge') # x, height
# plt.axhline(200, color='r', linestyle="--")
# plt.axvline(0.5, color='k')  # x=0.5 값에  수직 line 생성
plt.xticks(x, labels=['2017', '2018', '2019'])
plt.show()
```

```python
# stacked bar plot

import numpy as np
import matplotlib.pyplot as plt

menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

x = np.arange(5)

p1 = plt.bar(x, menMeans)
# p2 = plt.bar(x, womenMeans) #겹쳐서 안보임
p2 = plt.bar(x, womenMeans,  bottom=menMeans)

plt.ylabel('num')
plt.title('Exam')
plt.xticks(x, ('1', '2', '3', '4', '5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend(['Men', 'Women'])

plt.show()
```

```python
# stacked bar plot 2

import numpy as np
import matplotlib.pyplot as plt

catMeans = (20, 35, 30, 35, 27)
dogMeans = (25, 32, 34, 20, 25)
birdMeans = (5, 2, 4, 2, 5)

x = np.arange(5)

p1 = plt.bar(x, catMeans, width=0.35, color='#d62728')
p2 = plt.bar(x, dogMeans, width=0.35, bottom=catMeans)
p3 = plt.bar(x, birdMeans, width=0.35, bottom=catMeans)

plt.ylabel('num')
plt.title('Pet ')
plt.legend(['cat', 'dog','bird'])

plt.show()
```

## boxplot
- [공식레퍼런스](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html)

```python
## 가) 기본
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(1234)
# 박스plot는 홀수값 지정
x = np.random.random(9)
print(x)

print(sorted(x, reverse=True))
plt.boxplot(x,  labels=['xx_value'])
plt.show()

print("1사분위:", np.quantile(x, 0.25)) # 0.2764642551430967
print("2사분위:", np.quantile(x, 0.5))  # 0.6221087710398319
print("3사분위:", np.quantile(x, 0.75)) # 0.7853585837137692
Q1, Q3 = np.quantile(x, [0.25, 0.75])
IQR = Q3 - Q1
print("IQR:", IQR)
```

```python
# boxplt을 여러개 표출할 수 있다.
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(1234)
# 박스plot는 홀수값 지정
x = np.random.random(9)
print(x)

print(sorted(x, reverse=True))
plt.boxplot([x, x, x],  labels=['xx1_value', 'xx2_value', 'xx3_value'])
plt.show()
```

```python
## 나) df 적용
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(1234)

df = pd.DataFrame(np.random.rand(10, 5), columns=['A', 'B', 'C', 'D', 'E'])
print(df.describe())
plt.boxplot(df, labels=df.columns)
plt.show()
```

## 그래프에 text 추가
- plt.text(x, y, mesg, fonddict=None)

- [공식레퍼런스](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
# ax = plt.axes()
x = np.arange(3)
values = [100, 400, 900]

plt.bar(x, values)
plt.axhline(200, color='r', linestyle="--")
plt.xticks(x, labels=['2017', '2018', '2019'])

plt.text(0, 800, 'hello', fontsize='xx-large', fontstyle='oblique', color='blue')
# fontsize={'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
# fontstyle={'normal', 'italic', 'oblique'}

# text 지정
for idx, v in enumerate(values):
    # ax.text(x[idx], v, v)  # ax.text(x,y, mesg )
    plt.text(x[idx], v, v )  # plt.text(x,y, mesg )
plt.show()
```

## 그래프 파일 저장

- [plt.gcf](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.gcf.html)
- [savefig](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.savefig.html)

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
# ax = plt.axes()
x = np.arange(3)
values = [100, 400, 900]

plt.bar(x, values)
plt.axhline(200, color='r', linestyle="--")
plt.xticks(x, labels=['2017', '2018', '2019'])

plt.text(0, 800, 'hello', fontsize='xx-large', fontstyle='oblique', color='blue')
# fontsize={'xx-small', 'x-small', 'small', 'medium', 'large', 'x-large', 'xx-large'}
# fontstyle={'normal', 'italic', 'oblique'}

# text 지정
for idx, v in enumerate(values):
    # ax.text(x[idx], v, v)  # ax.text(x,y, mesg )
    plt.text(x[idx], v, v )  # plt.text(x,y, mesg )

x = plt.gcf() # 파일 저장

plt.show()

x.savefig('sample.png')
```

## Matplotlib 기본

```python
# 1. 시각화하기 위한 도화지 생성
fig = plt.figure(figsize=(8,4))  # figsize=( w, h ) 단위는 inch / 옵션
# 2. 도화지에 그리기 위한 붓 생성 ==> 거의 사용 안하고 plt로 직접 그래프 그림
ax = plt.axes()

# 그래프 그리기 => ax.그래프, plt.그래프(일반적인 방법)

# 그래프 보이기
plt.show() # py에서는 show를 명시해야 그래프가 보인다.
```

## 다중 그래프

```python
# 가) fig.add_subplot(행,열, n번째열)
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)  # 1행 2열의 1번째열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

ax2= fig.add_subplot(1,2,2)   # 1행 2열의 2번째열
plt.plot([-1,-2,-3,-4])
plt.title("sub2")

fig.suptitle("Main")
plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2,2,(1, 2))  # 1행 1열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub1")

# ax2= fig.add_subplot(2,2,2)   # 1행 2열
# plt.plot([-1,-2,-3,-4])
# plt.title("sub2")

ax3 = fig.add_subplot(2,2,3)  # 2행 1열
plt.plot([1,2,3,4])  # ax1.plot([1,2,3,4]) 대신에 plt 사용
plt.title("sub3")

ax4= fig.add_subplot(2,2,4)   # 2행 2열
plt.plot([-1,-2,-3,-4])
plt.title("sub4")

fig.suptitle("Main")
plt.show()
```

```python
# 나) plt.subplots(행,열)
import matplotlib.pyplot as plt
fig, axs = plt.subplots(nrows=1, ncols=2) # 1행 2열
print(axs)
axs[0].plot([1,2,3])
axs[1].plot([-1,-2,-3])
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2) # 2행 2열
print(axs, type(axs))
axs[0,0].plot([1,2,3])
axs[0,1].plot([-1,-2,-3])
axs[1,0].plot([1,2,3])
axs[1,1].plot([-1,-2,-3])
plt.show()
```

## 선 그래프

```python
# 1) y값만 지정
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))  # 생략가능

plt.plot([1,2,3,4])  # y 값 , x 값은 자동 생성
plt.show()
```

```python
# 2) x,y값 지정
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.show()
```

```python
# 3) x,y 라벨지정 
import matplotlib.pyplot as plt
# plt.rc('font', family='Malgun Gothic')
plt.rc('font', size=15)

fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.show()
```

```python
# 4) 축 범위 지정
import matplotlib.pyplot as plt

# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20]) # [xmin, xmax, ymin, ymax] 형태로 X, Y축의 범위를 지정한다.
print(plt.axis())  # (0.0, 5.0, 0.0, 20.0)
plt.show()
```

```python
# 5) 눈금 지정
import matplotlib.pyplot as plt
import numpy as np

# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 10, 0, 20]) # [xmin, xmax, ymin, ymax] 형태로 X, Y축의 범위를 지정한다.
# plt.xticks([0,5,10]) # x축 눈금지정
plt.xticks(range(11)) # x축 눈금지정
plt.yticks([0,1,5,10,20]) # y축 눈금지정
plt.show()
```

```python
# 6) 눈금 지정 + label
import matplotlib.pyplot as plt
import numpy as np

# fig = plt.figure(figsize=(8,6))  # 생략가능
plt.plot([1,2,3,4],[1, 4, 9, 16])  # x 값 , y 값
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
# plt.xticks([1,2,3,4], labels=['Jan',  'Feb',  'Mar',  'May'])
plt.xticks([1,2,3,4], labels=['Jan',  'Feb',  'Mar',  'May'], rotation=45)
plt.yticks([1, 4, 9, 16], ('0', '1GB', '2GB', '3GB'))
plt.show()
```

```python
# 7)색상지정
# https://matplotlib.org/stable/tutorials/colors/colors.html#specifying-colors
import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], color="b")
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 8)marker 지정
# https://matplotlib.org/stable/api/markers_api.html?highlight=marker#module-matplotlib.markers

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], "g*") #  o, s, p, ..
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 8) marker + 선 동시에 그리기
# https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], marker="*") #  o, s, p, ..
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 9) 선 스타일
# https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D.set_linestyle

import matplotlib.pyplot as plt
plt.plot([1,2,3,4],[1, 4, 9, 16], linestyle="-.", color='black') #  -, --, -. :
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.axis([0, 5, 0, 20])
plt.show()
```

```python
# 10) grid
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html

import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,2, 0.2)
plt.grid(True, axis='both', color='blue', linestyle=':') # axis = {'both', 'x', 'y'}
plt.plot(n,n, 'bo', label='linear')
plt.plot(n,n**2, 'r', marker="*", linewidth=2, label='quadratic')
plt.plot(n,n**3, 'y', marker="^", linewidth=3, label='cubic')
plt.legend()
plt.show()
```

## 여러 그래프 한 번에 그리기

```python
# 개별적으로 plot 생성 시 지정하거나 마지막에 plt.legend에 한 번에 넣어서 지정
# 개별적으로 생성해도 plt.legend를 실행해야 범례가 나온다.

import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0,2, 0.2)

plt.plot(n,n, 'bo', label='linear')
plt.plot(n,n**2, 'r', marker="*", linewidth=2, label='quadratic')
plt.plot(n,n**3, 'y', marker="^", linewidth=3, label='cubic')
plt.legend()
plt.show()
```
